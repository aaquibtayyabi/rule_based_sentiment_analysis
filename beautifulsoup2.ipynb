```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.insurancewatch.com.au/aia-insurance-reviews.html"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())

user_reviews= []
review_publish_date = []
reviewers= []

reviews = soup.find_all('span', itemprop='reviewBody')
for review in reviews:
    user_reviews.append(review.text)

dates = soup.find_all('meta', itemprop="datePublished")
for date in dates:
    review_publish_date.append(date['content'])

names = soup.find_all('span', itemprop='name')
for name in names:
    reviewers.append(name.text)

insurance_reviews = []
for review_final, date_final, name_final in zip(user_reviews, review_publish_date, reviewers):
    insurance_reviews.append([review_final, date_final, name_final])
print(insurance_reviews)


    
```

```python
df = pd.DataFrame(insurance_reviews, columns=["reviews", "dates", "names"])
```

```python
df.to_csv('aiareviews.csv')
```
